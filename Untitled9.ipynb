```python
# Data dari Playstore
df_play = pd.read_csv("/content/ulasan_djp_online_ribuan (1).csv")
df_play = df_play[['content']]
df_play['source'] = 'playstore'

# Data dari Instagram
df_ig = pd.read_csv("/content/komentar - komentar.csv", header=None, names=['content'])
df_ig['source'] = 'instagram'

# Gabungkan & hapus duplikat
df_all = pd.concat([df_play, df_ig], ignore_index=True).drop_duplicates()
df_all.dropna(subset=['content'], inplace=True)
```

```python
import pandas as pd
import re

# --- Load kamus kata baku ---
kamus_df = pd.read_excel('/content/kamuskatabaku.xlsx')
kamus = dict(zip(kamus_df['tidak_baku'], kamus_df['kata_baku']))

# --- Fungsi ubah ke kata baku ---
def konversi_kata_baku(tokens):
    return [kamus.get(token, token) for token in tokens]

# --- Fungsi hilangkan emoji ---
def remove_emoji(text):
    emoji_pattern = re.compile(
        "["u"\U0001F600-\U0001F64F"  # emoticon
        u"\U0001F300-\U0001F5FF"  # simbol & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # Chinese char
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

# --- Fungsi bersihkan teks dasar ---
def clean_text(text):
    text = remove_emoji(text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.lower()
    return text.strip()

# --- Tokenisasi ---
def tokenize(text):
    return text.split()

# --- Fungsi preprocessing utama ---
def preprocess_text(text):
    text = text.lower()
    text = remove_emoji(text)
    text = re.sub(r'[^a-z0-9\s]', '', text)
    tokens = text.split()
    tokens_baku = konversi_kata_baku(tokens)
    return tokens_baku

# --- Terapkan ke DataFrame ---
df_all['cleaned'] = df_all['content'].apply(clean_text)
df_all['t'] = df_all['content'].apply(preprocess_text)

# Cek hasil
df_all[['content', 'cleaned', 't']].head()
```

```python
import requests
import csv
from io import StringIO

# Lexicon positif
lexicon_positive = {}
response = requests.get('https://raw.githubusercontent.com/angelmetanosaa/dataset/main/lexicon_positive.csv')
if response.status_code == 200:
    reader = csv.reader(StringIO(response.text), delimiter=',')
    for row in reader:
        lexicon_positive[row[0].strip().lower()] = int(row[1])

response = requests.get('https://raw.githubusercontent.com/fajri91/InSet/master/positive.tsv')
if response.status_code == 200:
    reader = csv.reader(StringIO(response.text), delimiter='\t')
    next(reader)
    for row in reader:
        word = row[0].strip().lower()
        weight = int(row[1])
        lexicon_positive[word] = max(weight, lexicon_positive.get(word, 0))

# Lexicon negatif
lexicon_negative = {}
response = requests.get('https://raw.githubusercontent.com/angelmetanosaa/dataset/main/lexicon_negative.csv')
if response.status_code == 200:
    reader = csv.reader(StringIO(response.text), delimiter=',')
    for row in reader:
        lexicon_negative[row[0].strip().lower()] = int(row[1])

response = requests.get('https://raw.githubusercontent.com/fajri91/InSet/master/negative.tsv')
if response.status_code == 200:
    reader = csv.reader(StringIO(response.text), delimiter='\t')
    next(reader)
    for row in reader:
        word = row[0].strip().lower()
        weight = -abs(int(row[1]))
        lexicon_negative[word] = min(weight, lexicon_negative.get(word, 0))

# Tambahkan kata negatif
kata_negatif_manual = [
    'tidak', 'non', 'PALAK', 'bgst', 'kelamaan', 'lambat', 'perbaiki', 'BURUUUUKKKKK', 'cuapekkk', 'kenapa', 'kok', 'tapi', 'belum', 'susah', 'malah', 'begini',
    'kan', 'masih', 'ribet', 'gagal', 'error', 'eror', 'lupa', 'sampah', 'tidak bisa'
]

for kata in kata_negatif_manual:
    lexicon_negative[kata] = -9

# Gabungkan lexicon
lexicon_combined = {}
lexicon_combined.update(lexicon_positive)
lexicon_combined.update(lexicon_negative)
```

```python
# Gunakan hasil preprocessing untuk scoring
def score_sentiment_from_tokens(tokens):
    score = 0
    for word in tokens:
        score += lexicon_combined.get(word, 0)
    return score

df_all['sentiment_score'] = df_all['t'].apply(score_sentiment_from_tokens)

# Label sentimen fungsi
def label_sentiment(score):
    if score > 0:
        return 'positive'
    elif score < 0:
        return 'negative'
    else:
        return 'neutral'

df_all['sentiment'] = df_all['sentiment_score'].apply(label_sentiment)
```

# PEMETAAN MASALAH NEGATIF

```python
df_neg = df_all[df_all['sentiment'] == 'negative'].copy()
```

```python
df_neg.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

stop_words = set(stopwords.words('indonesian'))

def tokenize(text):
    tokens = [word for word in text.split() if word not in stop_words and len(word) > 2]
    return tokens

df_neg['tokens'] = df_neg['cleaned'].apply(tokenize)
df_neg['joined'] = df_neg['tokens'].apply(lambda x: ' '.join(x))
```

```python
vectorizer = CountVectorizer(max_df=0.9, min_df=5)
X = vectorizer.fit_transform(df_neg['joined'])
```

```python
from sklearn.decomposition import LatentDirichletAllocation

n_topics = 5
lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
lda.fit(X)
```

```python
def print_topics(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        top_words = [feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]
        print(f"\nüî∏ Topik #{topic_idx+1}:")
        print(", ".join(top_words))

print_topics(lda, vectorizer.get_feature_names_out(), 10)
```

```python
def plot_wordclouds(lda, vectorizer, n_top_words=10, n_topics=5):
    feature_names = vectorizer.get_feature_names_out()

    fig, axes = plt.subplots(1, n_topics, figsize=(20, 8), sharex=True)
    for topic_idx, topic in enumerate(lda.components_[:n_topics]):
        top_features_ind = topic.argsort()[:-n_top_words - 1:-1]
        top_features = [feature_names[i] for i in top_features_ind]
        weights = {feature_names[i]: topic[i] for i in top_features_ind}

        wordcloud = WordCloud(width=600, height=400,
                              background_color='white',
                              colormap='tab10').generate_from_frequencies(weights)
        axes[topic_idx].imshow(wordcloud, interpolation='bilinear')
        axes[topic_idx].axis("off")
        axes[topic_idx].set_title(f"Topik #{topic_idx+1}", fontsize=14)

    plt.tight_layout()
    plt.show()

plot_wordclouds(lda, vectorizer, n_top_words=10, n_topics=5)
```

#### TF-IDF klasifikasi sentimen

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
X = df_all['cleaned']
y = df_all['sentiment']
```

```python
# Ekstraksi fitur dengan TF-IDF
tfidf = TfidfVectorizer(max_features=200, min_df=17, max_df=0.8)
X_tfidf = tfidf.fit_transform(X)
```

```python
# Bagi data menjadi data latih dan uji
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)

print("Jumlah data latih:", X_train.shape[0])
print("Jumlah data uji:", X_test.shape[0])
```

```python
# Naive Bayes
nb_model = MultinomialNB()
nb_model.fit(X_train, y_train)
nb_pred = nb_model.predict(X_test)

# Logistic Regression
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)

# Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
```

```python
def evaluasi_model(y_true, y_pred, model_name):
    print(f"\n=== {model_name} ===")
    print("Akurasi:", accuracy_score(y_true, y_pred))
    print("Classification Report:")
    print(classification_report(y_true, y_pred))

    # Confusion Matrix
    # Changed labels to match the output of the label_sentiment function
    cm = confusion_matrix(y_true, y_pred, labels=["positive", "neutral", "negative"])
    # Changed xticklabels and yticklabels to match the actual labels
    sns.heatmap(cm, annot=True, fmt="d", xticklabels=["positive", "neutral", "negative"], yticklabels=["positive", "neutral", "negative"])
    plt.title(f"Confusion Matrix - {model_name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

# Evaluasi semua model
evaluasi_model(y_test, nb_pred, "Naive Bayes")
evaluasi_model(y_test, lr_pred, "Logistic Regression")
evaluasi_model(y_test, rf_pred, "Random Forest")
```
