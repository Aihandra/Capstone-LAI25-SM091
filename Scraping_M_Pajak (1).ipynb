```python
from google_play_scraper import reviews, Sort
import pandas as pd
import time
```

```python
# ====== Konfigurasi ======
app_id = 'id.go.pajak.djp'   # ID aplikasi DJP Online
count_per_request = 200      # Maksimum per permintaan
total_target = 10000         # Target ulasan maksimal
lang = 'id'                  # Bahasa
country = 'id'               # Negara
delay = 1                    # Delay antar request (dalam detik)
```

```python
# ====== Proses Scraping ======
all_reviews = []
```

```python
# Ambil batch pertama
result, continuation_token = reviews(
    app_id,
    lang=lang,
    country=country,
    sort=Sort.NEWEST,
    count=count_per_request
)
all_reviews.extend(result)
```

```python
# Tracking untuk mendeteksi stagnasi
last_count = 0
same_count_repeat = 0
max_repeat = 3

# Loop pengambilan data berikutnya
while continuation_token and len(all_reviews) < total_target:
    result, continuation_token = reviews(
        app_id,
        lang=lang,
        country=country,
        sort=Sort.NEWEST,
        count=count_per_request,
        continuation_token=continuation_token
    )
    all_reviews.extend(result)
    print(f"Total terkumpul: {len(all_reviews)}")

    # Deteksi stagnasi (kalau jumlah tidak berubah terus)
    if len(all_reviews) == last_count:
        same_count_repeat += 1
    else:
        same_count_repeat = 0
        last_count = len(all_reviews)

    if same_count_repeat >= max_repeat:
        print("üîÅ Tidak ada penambahan data baru, scraping dihentikan otomatis.")
        break

    time.sleep(delay)
```

```python
df = pd.DataFrame(all_reviews)
```

```python
df.head(20)
```

```python
# ====== Pembersihan Teks ======
stopword_factory = StopWordRemoverFactory()
stopwords = set(stopword_factory.get_stop_words())
stemmer = StemmerFactory().create_stemmer()
```

```python
def clean_text(text):
    if not isinstance(text, str):
        return ''
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = re.sub(r'\s+', ' ', text).strip()
    text = ' '.join([word for word in text.split() if word not in stopwords])
    text = stemmer.stem(text)
    return text
```

```python
# Tambahkan kolom hasil pembersihan
df['clean_content'] = df['content'].apply(clean_text)
```
